## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
data = pd.read_csv('data/hist_data.csv')
plt.hist(data['aデータ'], bins=30, alpha=0.5, label='aデータ', color='orange')
plt.hist(data['bデータ'], bins=30, alpha=0.5, label='bデータ', color='turquoise')
plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
data = pd.read_csv('data/scatter_data.csv')

class0 = data[data['class'] == 0]
class1 = data[data['class'] == 1]
class2 = data[data['class'] == 2]

plt.scatter(class0['length (cm)'], class0['width (cm)'], color='red', label='class0')
plt.scatter(class1['length (cm)'], class1['width (cm)'], color='blue', label='class1')
plt.scatter(class2['length (cm)'], class2['width (cm)'], color='green', label='class2')

plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.legend()
plt.grid(True)

plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df = pd.read_csv('data/line_data.csv')

plt.figure(figsize=(10, 6)) 
plt.plot(df['期間'], df['件数'], marker='o')
plt.xlabel('期間')
plt.ylabel('件数）')
plt.title('期間ごとの件数の推移')
plt.grid(True)  
plt.xticks(rotation=45) 
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(df['期間'], df['費用'])
plt.xlabel('期間')
plt.ylabel('費用（円）')
plt.title('期間ごとの費用の分布')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
